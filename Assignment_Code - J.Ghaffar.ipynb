```python
# Self-Organizing Maps (SOMs) for Dimensionality Reduction and Visualization
# Author: J.Ghaffar Student#23104340

# This notebook demonstrates the implementation and application of Self-Organizing Maps
# for dimensionality reduction and visualization of high-dimensional data, with a focus
# on how different neighborhood functions affect the topology preservation.

# MIT License

# Copyright (c) 2023 [J.Ghaffar]

# Permission is hereby granted, free of charge, to any person obtaining a copy
# of this software and associated documentation files (the "Software"), to deal
# in the Software without restriction, including without limitation the rights
# to use, copy, modify, merge, publish, distribute, sublicense, and/or sell
# copies of the Software, and to permit persons to whom the Software is
# furnished to do so, subject to the following conditions:

# The above copyright notice and this permission notice shall be included in all
# copies or substantial portions of the Software.

# THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR
# IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY,
# FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL THE
# AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER
# LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING FROM,
# OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER DEALINGS IN THE
# SOFTWARE.

# To run the script:
# 1. Ensure you have the required libraries installed (numpy, matplotlib, pandas,
#    scikit-learn, seaborn, tqdm).
# 2. Execute the script in a Python environment (e.g., Jupyter Notebook, Google Colab).
# 3. The script will load datasets, train SOM models, and generate visualizations.
# 4. Modify parameters as desired to explore different SOM configurations.

# For detailed explanations and code walkthrough, refer to the comments within the code.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris, fetch_olivetti_faces, make_blobs
from sklearn.manifold import TSNE
from matplotlib.colors import ListedColormap
import seaborn as sns
from tqdm.notebook import tqdm

# Set random seed for reproducibility
np.random.seed(42)

# Part 1: Implementation of Self-Organizing Maps from scratch

class SelfOrganizingMap:
    """
    Implementation of a Self-Organizing Map (SOM) for dimensionality reduction and visualization.

    Parameters:
    -----------
    map_size : tuple
        The size of the SOM grid (height, width)
    input_dim : int
        The dimensionality of the input data
    learning_rate : float
        Initial learning rate
    sigma : float
        Initial neighborhood radius
    neighborhood_function : str
        Type of neighborhood function ('gaussian', 'mexican_hat', 'bubble')
    decay_function : str
        Type of learning rate decay ('linear', 'exponential')
    n_iterations : int
        Number of iterations for training
    """

    def __init__(self, map_size=(10, 10), input_dim=2, learning_rate=0.5, sigma=1.0,
                 neighborhood_function='gaussian', decay_function='linear', n_iterations=1000):
        self.map_size = map_size
        self.input_dim = input_dim
        self.initial_learning_rate = learning_rate
        self.learning_rate = learning_rate
        self.initial_sigma = sigma
        self.sigma = sigma
        self.n_iterations = n_iterations
        self.current_iteration = 0
        self.neighborhood_function = neighborhood_function
        self.decay_function = decay_function

        # Initialize the SOM weight vectors randomly
        self.weights = np.random.rand(map_size[0], map_size[1], input_dim)

        # Pre-compute the coordinates of each neuron in the grid
        self.neuron_coordinates = np.array(
            [[i, j] for i in range(map_size[0]) for j in range(map_size[1])]
        ).reshape(map_size[0], map_size[1], 2)

    def _find_bmu(self, x):
        """
        Find the Best Matching Unit (BMU) for the input vector x

        Parameters:
        -----------
        x : numpy.ndarray
            Input vector

        Returns:
        --------
        tuple
            Coordinates of the BMU (row, column)
        """
        # Calculate the distance between x and all weight vectors
        distances = np.sum((self.weights - x) ** 2, axis=2)

        # Find the coordinates of the minimum distance
        bmu_idx = np.unravel_index(np.argmin(distances), distances.shape)
        return bmu_idx

    def _decay_learning_rate(self):
        """Update the learning rate according to the selected decay function"""
        if self.decay_function == 'linear':
            self.learning_rate = self.initial_learning_rate * (1 - self.current_iteration / self.n_iterations)
        elif self.decay_function == 'exponential':
            self.learning_rate = self.initial_learning_rate * np.exp(-self.current_iteration / self.n_iterations)

    def _decay_sigma(self):
        """Update the neighborhood radius according to a linear decay"""
        self.sigma = self.initial_sigma * (1 - self.current_iteration / self.n_iterations)

    def _calculate_influence(self, bmu_idx, decay=True):
        """
        Calculate the influence of the BMU on each neuron in the grid

        Parameters:
        -----------
        bmu_idx : tuple
            Coordinates of the BMU (row, column)
        decay : bool
            Whether to decay the sigma value

        Returns:
        --------
        numpy.ndarray
            Influence matrix with the same shape as the SOM grid
        """
        if decay:
            self._decay_sigma()

        # Get the coordinates of the BMU
        bmu_location = np.array([bmu_idx[0], bmu_idx[1]])

        # Calculate the squared distance of each neuron from the BMU
        distances = np.sum((self.neuron_coordinates - bmu_location) ** 2, axis=2)

        # Apply the selected neighborhood function
        if self.neighborhood_function == 'gaussian':
            # Gaussian neighborhood function
            influence = np.exp(-distances / (2 * self.sigma ** 2))
        elif self.neighborhood_function == 'mexican_hat':
            # Mexican hat (Ricker wavelet) neighborhood function
            # This function has a central positive region surrounded by a negative region
            r_squared = distances / (self.sigma ** 2)
            influence = (2 - r_squared) * np.exp(-r_squared / 2)
        elif self.neighborhood_function == 'bubble':
            # Bubble (step) neighborhood function
            influence = (distances <= self.sigma ** 2).astype(float)
        else:
            raise ValueError(f"Unknown neighborhood function: {self.neighborhood_function}")

        return influence

    def train_single_step(self, x):
        """
        Train the SOM on a single input vector

        Parameters:
        -----------
        x : numpy.ndarray
            Input vector
        """
        # Find the BMU
        bmu_idx = self._find_bmu(x)

        # Calculate the influence of the BMU on each neuron
        influence = self._calculate_influence(bmu_idx)

        # Update the learning rate
        self._decay_learning_rate()

        # Update the weights of each neuron
        # Reshape influence to match weights dimensions for broadcasting
        influence_reshaped = influence.reshape(self.map_size[0], self.map_size[1], 1)

        # Update formula: w(t+1) = w(t) + learning_rate * influence * (x - w(t))
        self.weights += self.learning_rate * influence_reshaped * (x - self.weights)

        # Increment the iteration counter
        self.current_iteration += 1

    def train(self, data, verbose=True):
        """
        Train the SOM on the entire dataset

        Parameters:
        -----------
        data : numpy.ndarray
            Training data with shape (n_samples, input_dim)
        verbose : bool
            Whether to show progress bar
        """
        # Reset iteration counter
        self.current_iteration = 0

        # Train for the specified number of iterations
        iterator = tqdm(range(self.n_iterations)) if verbose else range(self.n_iterations)

        for _ in iterator:
            # Randomly select a data point
            x = data[np.random.randint(0, data.shape[0])]
            self.train_single_step(x)

    def transform(self, data):
        """
        Transform data to grid coordinates by finding the BMU for each data point

        Parameters:
        -----------
        data : numpy.ndarray
            Data to transform with shape (n_samples, input_dim)

        Returns:
        --------
        numpy.ndarray
            Grid coordinates for each data point with shape (n_samples, 2)
        """
        bmu_coordinates = np.zeros((data.shape[0], 2))

        for i, x in enumerate(data):
            bmu_idx = self._find_bmu(x)
            bmu_coordinates[i] = np.array(bmu_idx)

        return bmu_coordinates

    def get_umatrix(self):
        """
        Compute the U-Matrix (Unified Distance Matrix) which shows the distances
        between neighboring neurons

        Returns:
        --------
        numpy.ndarray
            U-Matrix with shape (map_size[0], map_size[1])
        """
        umatrix = np.zeros(self.map_size)

        for i in range(self.map_size[0]):
            for j in range(self.map_size[1]):
                # Get the weight vector of the current neuron
                w_ij = self.weights[i, j]

                # Initialize sum of distances and count of neighbors
                sum_dist = 0
                count = 0

                # Check all 8 neighboring neurons (if they exist)
                for di in [-1, 0, 1]:
                    for dj in [-1, 0, 1]:
                        # Skip the current neuron
                        if di == 0 and dj == 0:
                            continue

                        # Check if the neighbor exists
                        ni, nj = i + di, j + dj
                        if 0 <= ni < self.map_size[0] and 0 <= nj < self.map_size[1]:
                            # Calculate the Euclidean distance between weight vectors
                            w_neighbor = self.weights[ni, nj]
                            dist = np.linalg.norm(w_ij - w_neighbor)
                            sum_dist += dist
                            count += 1

                # Average distance to neighbors
                if count > 0:
                    umatrix[i, j] = sum_dist / count

        return umatrix

    def visualize_umatrix(self, ax=None, cmap='viridis', title="U-Matrix"):
        """
        Visualize the U-Matrix

        Parameters:
        -----------
        ax : matplotlib.axes.Axes, optional
            Matplotlib axes to plot on
        cmap : str, optional
            Colormap to use
        title : str, optional
            Title of the plot

        Returns:
        --------
        matplotlib.axes.Axes
            The axes object with the plot
        """
        if ax is None:
            fig, ax = plt.subplots(figsize=(10, 8))

        umatrix = self.get_umatrix()
        im = ax.imshow(umatrix, cmap=cmap)
        plt.colorbar(im, ax=ax)
        ax.set_title(title)

        return ax

    def visualize_weight_planes(self, feature_names=None, figsize=(15, 12)):
        """
        Visualize the weight planes for each input dimension

        Parameters:
        -----------
        feature_names : list, optional
            Names of the input features
        figsize : tuple, optional
            Size of the figure

        Returns:
        --------
        matplotlib.figure.Figure
            The figure with the plots
        """
        # Determine number of features to visualize
        n_features = self.input_dim

        # Calculate grid dimensions for subplots
        n_cols = min(3, n_features)
        n_rows = int(np.ceil(n_features / n_cols))

        # Create figure and subplots
        fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)

        # Flatten axes array for easy iteration
        axes = axes.flatten() if n_features > 1 else [axes]

        # Plot each weight component
        for i in range(n_features):
            # Extract weights for this feature
            feature_weights = self.weights[:, :, i]

            # Plot weights
            im = axes[i].imshow(feature_weights, cmap='coolwarm')
            plt.colorbar(im, ax=axes[i])

            # Set title
            feature_name = feature_names[i] if feature_names is not None else f"Feature {i+1}"
            axes[i].set_title(f"Weight plane: {feature_name}")

        # Hide any unused subplots
        for i in range(n_features, len(axes)):
            axes[i].axis('off')

        plt.tight_layout()
        return fig

    def visualize_sample_mapping(self, data, labels=None, ax=None, cmap='viridis',
                                s=100, alpha=0.7, title="SOM Mapping"):
        """
        Visualize how samples map to the SOM grid

        Parameters:
        -----------
        data : numpy.ndarray
            Data samples with shape (n_samples, input_dim)
        labels : numpy.ndarray, optional
            Labels for data points
        ax : matplotlib.axes.Axes, optional
            Matplotlib axes to plot on
        cmap : str or matplotlib.colors.Colormap, optional
            Colormap to use
        s : float, optional
            Marker size
        alpha : float, optional
            Transparency of markers
        title : str, optional
            Title of the plot

        Returns:
        --------
        matplotlib.axes.Axes
            The axes object with the plot
        """
        if ax is None:
            fig, ax = plt.subplots(figsize=(10, 8))

        # Get BMU coordinates for each data point
        bmu_coordinates = self.transform(data)

        # Plot U-Matrix as background
        umatrix = self.get_umatrix()
        im = ax.imshow(umatrix, cmap='Greys', alpha=0.6)

        # Add colorbar for U-Matrix
        cbar = plt.colorbar(im, ax=ax)
        cbar.set_label('Average distance to neighbors')

        # Plot data points on top
        if labels is not None:
            # Get unique labels
            unique_labels = np.unique(labels)

            # Create a colormap with a unique color for each label
            n_labels = len(unique_labels)
            colors = plt.cm.get_cmap(cmap, n_labels)

            # Plot each label group
            for i, label in enumerate(unique_labels):
                mask = labels == label
                ax.scatter(bmu_coordinates[mask, 1], bmu_coordinates[mask, 0],
                          c=[colors(i)], label=f"Class {label}", s=s, alpha=alpha)

            ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        else:
            # Plot all points with the same color
            ax.scatter(bmu_coordinates[:, 1], bmu_coordinates[:, 0], c='blue', s=s, alpha=alpha)

        ax.set_title(title)
        ax.set_xlim(-0.5, self.map_size[1] - 0.5)
        ax.set_ylim(-0.5, self.map_size[0] - 0.5)
        ax.invert_yaxis()  # Invert y-axis to match imshow orientation

        return ax

# Part 2: Exploring Self-Organizing Maps with the Iris dataset

# Load the Iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
feature_names_iris = iris.feature_names

print("Iris dataset shape:", X_iris.shape)
print("Number of classes:", len(np.unique(y_iris)))
print("Feature names:", feature_names_iris)

# Standardize the data
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

# Initialize and train a SOM with gaussian neighborhood function
som_gaussian = SelfOrganizingMap(
    map_size=(10, 10),
    input_dim=4,
    learning_rate=0.5,
    sigma=2.0,
    neighborhood_function='gaussian',
    n_iterations=5000
)

print("Training SOM with Gaussian neighborhood function...")
som_gaussian.train(X_iris_scaled)

# Initialize and train a SOM with mexican hat neighborhood function
som_mexican = SelfOrganizingMap(
    map_size=(10, 10),
    input_dim=4,
    learning_rate=0.5,
    sigma=2.0,
    neighborhood_function='mexican_hat',
    n_iterations=5000
)

print("Training SOM with Mexican Hat neighborhood function...")
som_mexican.train(X_iris_scaled)

# Initialize and train a SOM with bubble neighborhood function
som_bubble = SelfOrganizingMap(
    map_size=(10, 10),
    input_dim=4,
    learning_rate=0.5,
    sigma=2.0,
    neighborhood_function='bubble',
    n_iterations=5000
)

print("Training SOM with Bubble neighborhood function...")
som_bubble.train(X_iris_scaled)

# Compare the three SOMs
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Visualize U-Matrix for each SOM
som_gaussian.visualize_umatrix(ax=axes[0], title="U-Matrix (Gaussian)")
som_mexican.visualize_umatrix(ax=axes[1], title="U-Matrix (Mexican Hat)")
som_bubble.visualize_umatrix(ax=axes[2], title="U-Matrix (Bubble)")

plt.tight_layout()
plt.show()

# Visualize sample mapping for each SOM
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

som_gaussian.visualize_sample_mapping(X_iris_scaled, y_iris, ax=axes[0],
                                     title="Sample Mapping (Gaussian)")
som_mexican.visualize_sample_mapping(X_iris_scaled, y_iris, ax=axes[1],
                                    title="Sample Mapping (Mexican Hat)")
som_bubble.visualize_sample_mapping(X_iris_scaled, y_iris, ax=axes[2],
                                   title="Sample Mapping (Bubble)")

plt.tight_layout()
plt.show()

# Visualize weight planes for the Gaussian SOM
som_gaussian.visualize_weight_planes(feature_names=feature_names_iris)
plt.tight_layout()
plt.show()

# Part 3: Compare SOM with t-SNE for dimensionality reduction

# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
X_iris_tsne = tsne.fit_transform(X_iris_scaled)

# Plot t-SNE results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_iris_tsne[:, 0], X_iris_tsne[:, 1], c=y_iris,
                     cmap='viridis', s=100, alpha=0.7)
plt.title("t-SNE visualization of Iris dataset")
plt.colorbar(scatter, label="Class")
plt.show()

# Part 4: Advanced SOM visualization with a more complex dataset

# Load the Olivetti faces dataset
faces = fetch_olivetti_faces()
X_faces = faces.data
y_faces = faces.target

print("Olivetti faces dataset shape:", X_faces.shape)
print("Number of individuals:", len(np.unique(y_faces)))

# Standardize the data
X_faces_scaled = scaler.fit_transform(X_faces)

# Train a SOM with the faces dataset
som_faces = SelfOrganizingMap(
    map_size=(15, 15),
    input_dim=X_faces.shape[1],
    learning_rate=0.5,
    sigma=3.0,
    neighborhood_function='gaussian',
    n_iterations=10000
)

print("Training SOM with the Olivetti faces dataset...")
som_faces.train(X_faces_scaled)

# Visualize the sample mapping
plt.figure(figsize=(12, 10))
som_faces.visualize_sample_mapping(X_faces_scaled, y_faces,
                                  title="SOM Mapping of Olivetti Faces",
                                  cmap='tab20')
plt.show()

# Part 5: Demonstrate how SOM preserves topological relationships

# Create a synthetic dataset with a specific structure
n_samples = 1000
centers = [(0, 0, 0), (5, 5, 0), (0, 5, 5), (5, 0, 5)]
X_blobs, y_blobs = make_blobs(n_samples=n_samples, n_features=3, centers=centers,
                             cluster_std=0.8, random_state=42)

# Visualize the 3D dataset
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_blobs[:, 0], X_blobs[:, 1], X_blobs[:, 2],
                    c=y_blobs, cmap='viridis', s=50, alpha=0.7)
plt.title("3D Synthetic Dataset")
plt.colorbar(scatter, label="Cluster")
plt.show()

# Train SOMs with different neighborhood functions on the 3D dataset
# Standardize the data
X_blobs_scaled = scaler.fit_transform(X_blobs)

# Train SOMs with different neighborhood functions
som_gaussian_3d = SelfOrganizingMap(
    map_size=(10, 10),
    input_dim=3,
    learning_rate=0.5,
    sigma=2.0,
    neighborhood_function='gaussian',
    n_iterations=5000
)

som_mexican_3d = SelfOrganizingMap(
    map_size=(10, 10),
    input_dim=3,
    learning_rate=0.5,
    sigma=2.0,
    neighborhood_function='mexican_hat',
    n_iterations=5000
)

som_bubble_3d = SelfOrganizingMap(
    map_size=(10, 10),
    input_dim=3,
    learning_rate=0.5,
    sigma=2.0,
    neighborhood_function='bubble',
    n_iterations=5000
)

print("Training SOMs with different neighborhood functions on 3D data...")
som_gaussian_3d.train(X_blobs_scaled)
som_mexican_3d.train(X_blobs_scaled)
som_bubble_3d.train(X_blobs_scaled)

# Visualize the sample mappings
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

som_gaussian_3d.visualize_sample_mapping(X_blobs_scaled, y_blobs, ax=axes[0],
                                       title="Gaussian Neighborhood")
som_mexican_3d.visualize_sample_mapping(X_blobs_scaled, y_blobs, ax=axes[1],
                                      title="Mexican Hat Neighborhood")
som_bubble_3d.visualize_sample_mapping(X_blobs_scaled, y_blobs, ax=axes[2],
                                     title="Bubble Neighborhood")

plt.tight_layout()
plt.show()

# Part 6: Measure and compare the quality of the different SOM models

def calculate_topographic_error(som, data):
    """
    Calculate the topographic error of a SOM.

    The topographic error is the proportion of data points for which the first and second
    best matching units are not adjacent in the grid.

    Parameters:
    -----------
    som : SelfOrganizingMap
        Trained SOM
    data : numpy.ndarray
        Data samples with shape (n_samples, input_dim)

    Returns:
    --------
    float
        Topographic error (between 0 and 1)
    """
    errors = 0

    for x in data:
        # Calculate distances to all weight vectors
        distances = np.sum((som.weights - x) ** 2, axis=2)

        # Find the indices of the two best matching units
        flat_indices = np.argsort(distances.flatten())
        bmu_idx = np.unravel_index(flat_indices[0], distances.shape)
        second_bmu_idx = np.unravel_index(flat_indices[1], distances.shape)

        # Check if the two BMUs are adjacent
        # Two neurons are adjacent if the Manhattan distance between them is 1
        manhattan_dist = abs(bmu_idx[0] - second_bmu_idx[0]) + abs(bmu_idx[1] - second_bmu_idx[1])

        if manhattan_dist > 1:
            errors += 1

    return errors / data.shape[0]

def calculate_quantization_error(som, data):
    """
    Calculate the quantization error of a SOM.

    The quantization error is the average distance between each data point and its BMU.

    Parameters:
    -----------
    som : SelfOrganizingMap
        Trained SOM
    data : numpy.ndarray
        Data samples with shape (n_samples, input_dim)

    Returns:
    --------
    float
        Quantization error
    """
    error_sum = 0

    for x in data:
        # Find the BMU
        bmu_idx = som._find_bmu(x)

        # Calculate the distance between x and the BMU's weight vector
        bmu_weight = som.weights[bmu_idx]
        error = np.linalg.norm(x - bmu_weight)
        error_sum += error

    return error_sum / data.shape[0]

# Calculate and compare the errors for different neighborhood functions on the Iris dataset
topographic_error_gaussian = calculate_topographic_error(som_gaussian, X_iris_scaled)
topographic_error_mexican = calculate_topographic_error(som_mexican, X_iris_scaled)
topographic_error_bubble = calculate_topographic_error(som_bubble, X_iris_scaled)

quantization_error_gaussian = calculate_quantization_error(som_gaussian, X_iris_scaled)
quantization_error_mexican = calculate_quantization_error(som_mexican, X_iris_scaled)
quantization_error_bubble = calculate_quantization_error(som_bubble, X_iris_scaled)

print("\nIris Dataset Error Metrics:")
print(f"Gaussian Neighborhood - Topographic Error: {topographic_error_gaussian:.4f}, Quantization Error: {quantization_error_gaussian:.4f}")
print(f"Mexican Hat Neighborhood - Topographic Error: {topographic_error_mexican:.4f}, Quantization Error: {quantization_error_mexican:.4f}")
print(f"Bubble Neighborhood - Topographic Error: {topographic_error_bubble:.4f}, Quantization Error: {quantization_error_bubble:.4f}")

# Calculate and compare the errors for different neighborhood functions on the 3D dataset
topographic_error_gaussian_3d = calculate_topographic_error(som_gaussian_3d, X_blobs_scaled)
topographic_error_mexican_3d = calculate_topographic_error(som_mexican_3d, X_blobs_scaled)
topographic_error_bubble_3d = calculate_topographic_error(som_bubble_3d, X_blobs_scaled)

quantization_error_gaussian_3d = calculate_quantization_error(som_gaussian_3d, X_blobs_scaled)
quantization_error_mexican_3d = calculate_quantization_error(som_mexican_3d, X_blobs_scaled)
quantization_error_bubble_3d = calculate_quantization_error(som_bubble_3d, X_blobs_scaled)

print("\n3D Dataset Error Metrics:")
print(f"Gaussian Neighborhood - Topographic Error: {topographic_error_gaussian_3d:.4f}, Quantization Error: {quantization_error_gaussian_3d:.4f}")
print(f"Mexican Hat Neighborhood - Topographic Error: {topographic_error_mexican_3d:.4f}, Quantization Error: {quantization_error_mexican_3d:.4f}")
print(f"Bubble Neighborhood - Topographic Error: {topographic_error_bubble_3d:.4f}, Quantization Error: {quantization_error_bubble_3d:.4f}")

# Visualize the error metrics as a bar chart
metrics = ['Topographic Error (Iris)', 'Quantization Error (Iris)',
          'Topographic Error (3D)', 'Quantization Error (3D)']
gaussian_errors = [topographic_error_gaussian, quantization_error_gaussian,
                  topographic_error_gaussian_3d, quantization_error_gaussian_3d]
mexican_errors = [topographic_error_mexican, quantization_error_mexican,
                 topographic_error_mexican_3d, quantization_error_mexican_3d]
bubble_errors = [topographic_error_bubble, quantization_error_bubble,
                topographic_error_bubble_3d, quantization_error_bubble_3d]

plt.figure(figsize=(12, 8))
x = np.arange(len(metrics))
width = 0.25

plt.bar(x - width, gaussian_errors, width, label='Gaussian')
plt.bar(x, mexican_errors, width, label='Mexican Hat')
plt.bar(x + width, bubble_errors, width, label='Bubble')

plt.xlabel('Metric')
plt.ylabel('Error Value')
plt.title('SOM Error Metrics Comparison')
plt.xticks(x, metrics)
plt.legend()
plt.tight_layout()
plt.show()

# Part 7: Visualize the impact of different learning rates and neighborhood radii

# Function to train a SOM with specific parameters and return error metrics
def train_som_and_evaluate(data, map_size, input_dim, learning_rate, sigma,
                          neighborhood_function, n_iterations):
    """
    Train a SOM with the given parameters and evaluate its performance

    Returns:
    --------
    tuple
        (som, topographic_error, quantization_error)
    """
    som = SelfOrganizingMap(
        map_size=map_size,
        input_dim=input_dim,
        learning_rate=learning_rate,
        sigma=sigma,
        neighborhood_function=neighborhood_function,
        n_iterations=n_iterations
    )

    som.train(data, verbose=False)

    topographic_error = calculate_topographic_error(som, data)
    quantization_error = calculate_quantization_error(som, data)

    return som, topographic_error, quantization_error

# Test different learning rates
learning_rates = [0.1, 0.3, 0.5, 0.7, 0.9]
topographic_errors_lr = []
quantization_errors_lr = []

print("\nTesting different learning rates...")
for lr in learning_rates:
    _, te, qe = train_som_and_evaluate(
        X_iris_scaled, (10, 10), 4, lr, 2.0, 'gaussian', 5000
    )
    topographic_errors_lr.append(te)
    quantization_errors_lr.append(qe)
    print(f"Learning rate {lr}: Topographic Error = {te:.4f}, Quantization Error = {qe:.4f}")

# Test different sigma values
# Continuing from previous code...

# Test different sigma values
sigmas = [0.5, 1.0, 2.0, 3.0, 5.0]
topographic_errors_sigma = []
quantization_errors_sigma = []

print("\nTesting different sigma values...")
for sigma in sigmas:
    _, te, qe = train_som_and_evaluate(
        X_iris_scaled, (10, 10), 4, 0.5, sigma, 'gaussian', 5000
    )
    topographic_errors_sigma.append(te)
    quantization_errors_sigma.append(qe)
    print(f"Sigma {sigma}: Topographic Error = {te:.4f}, Quantization Error = {qe:.4f}")

# Plot the effect of learning rate and sigma on error metrics
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot topographic error vs learning rate
axes[0, 0].plot(learning_rates, topographic_errors_lr, 'o-', color='blue')
axes[0, 0].set_xlabel('Learning Rate')
axes[0, 0].set_ylabel('Topographic Error')
axes[0, 0].set_title('Topographic Error vs Learning Rate')
axes[0, 0].grid(True)

# Plot quantization error vs learning rate
axes[0, 1].plot(learning_rates, quantization_errors_lr, 'o-', color='red')
axes[0, 1].set_xlabel('Learning Rate')
axes[0, 1].set_ylabel('Quantization Error')
axes[0, 1].set_title('Quantization Error vs Learning Rate')
axes[0, 1].grid(True)

# Plot topographic error vs sigma
axes[1, 0].plot(sigmas, topographic_errors_sigma, 'o-', color='blue')
axes[1, 0].set_xlabel('Sigma')
axes[1, 0].set_ylabel('Topographic Error')
axes[1, 0].set_title('Topographic Error vs Sigma')
axes[1, 0].grid(True)

# Plot quantization error vs sigma
axes[1, 1].plot(sigmas, quantization_errors_sigma, 'o-', color='red')
axes[1, 1].set_xlabel('Sigma')
axes[1, 1].set_ylabel('Quantization Error')
axes[1, 1].set_title('Quantization Error vs Sigma')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# Part 8: Examine the convergence of the SOM training process

# Train a SOM and record the quantization error at each iteration
def train_som_with_error_tracking(data, map_size, input_dim, learning_rate, sigma,
                                neighborhood_function, n_iterations, track_interval=100):
    """
    Train a SOM and track the quantization error over iterations

    Returns:
    --------
    tuple
        (som, error_iterations, quantization_errors)
    """
    som = SelfOrganizingMap(
        map_size=map_size,
        input_dim=input_dim,
        learning_rate=learning_rate,
        sigma=sigma,
        neighborhood_function=neighborhood_function,
        n_iterations=n_iterations
    )

    # Reset iteration counter
    som.current_iteration = 0

    # Initialize error tracking
    error_iterations = []
    quantization_errors = []

    # Train and track errors
    for i in tqdm(range(n_iterations)):
        # Randomly select a data point
        x = data[np.random.randint(0, data.shape[0])]
        som.train_single_step(x)

        # Track error at specified intervals
        if i % track_interval == 0 or i == n_iterations - 1:
            error_iterations.append(i)
            quantization_errors.append(calculate_quantization_error(som, data))

    return som, error_iterations, quantization_errors

# Track convergence for different neighborhood functions
n_iterations = 10000
track_interval = 200

print("\nTracking convergence for Gaussian neighborhood function...")
_, iterations_gaussian, errors_gaussian = train_som_with_error_tracking(
    X_iris_scaled, (10, 10), 4, 0.5, 2.0, 'gaussian', n_iterations, track_interval
)

print("Tracking convergence for Mexican Hat neighborhood function...")
_, iterations_mexican, errors_mexican = train_som_with_error_tracking(
    X_iris_scaled, (10, 10), 4, 0.5, 2.0, 'mexican_hat', n_iterations, track_interval
)

print("Tracking convergence for Bubble neighborhood function...")
_, iterations_bubble, errors_bubble = train_som_with_error_tracking(
    X_iris_scaled, (10, 10), 4, 0.5, 2.0, 'bubble', n_iterations, track_interval
)

# Plot the convergence curves
plt.figure(figsize=(12, 8))
plt.plot(iterations_gaussian, errors_gaussian, 'o-', label='Gaussian')
plt.plot(iterations_mexican, errors_mexican, 's-', label='Mexican Hat')
plt.plot(iterations_bubble, errors_bubble, '^-', label='Bubble')
plt.xlabel('Iteration')
plt.ylabel('Quantization Error')
plt.title('Convergence of SOM Training')
plt.legend()
plt.grid(True)
plt.show()

# Part 9: Visualization of SOM weight adaptation during training

def visualize_som_training(data, map_size, input_dim, learning_rate, sigma,
                          neighborhood_function, n_iterations, sample_intervals):
    """
    Visualize the SOM weight adaptation during training

    Parameters:
    -----------
    data : numpy.ndarray
        Training data
    map_size : tuple
        Size of the SOM grid
    input_dim : int
        Dimensionality of the input data
    learning_rate : float
        Initial learning rate
    sigma : float
        Initial neighborhood radius
    neighborhood_function : str
        Type of neighborhood function
    n_iterations : int
        Total number of iterations
    sample_intervals : list
        List of iterations at which to sample the SOM state

    Returns:
    --------
    list
        List of trained SOMs at each sampled iteration
    """
    soms = []

    for n_iter in sample_intervals:
        print(f"Training SOM for {n_iter} iterations...")
        som = SelfOrganizingMap(
            map_size=map_size,
            input_dim=input_dim,
            learning_rate=learning_rate,
            sigma=sigma,
            neighborhood_function=neighborhood_function,
            n_iterations=n_iter
        )
        som.train(data, verbose=False)
        soms.append(som)

    return soms

# Sample SOM states at different iterations
sample_intervals = [10, 100, 500, 1000, 5000]
soms_gaussian = visualize_som_training(
    X_iris_scaled, (10, 10), 4, 0.5, 2.0, 'gaussian', 5000, sample_intervals
)

# Visualize U-matrices at different training stages
fig, axes = plt.subplots(1, len(sample_intervals), figsize=(20, 4))

for i, (som, n_iter) in enumerate(zip(soms_gaussian, sample_intervals)):
    som.visualize_umatrix(ax=axes[i], title=f"After {n_iter} iterations")

plt.tight_layout()
plt.show()

# Visualize sample mappings at different training stages
fig, axes = plt.subplots(1, len(sample_intervals), figsize=(20, 4))

for i, (som, n_iter) in enumerate(zip(soms_gaussian, sample_intervals)):
    som.visualize_sample_mapping(X_iris_scaled, y_iris, ax=axes[i],
                               title=f"After {n_iter} iterations")

plt.tight_layout()
plt.show()

# Part 10: Comparison of different map sizes and their effect on visualization

def train_soms_with_different_sizes(data, input_dim, map_sizes, learning_rate, sigma,
                                  neighborhood_function, n_iterations):
    """
    Train SOMs with different map sizes

    Returns:
    --------
    list
        List of trained SOMs with different map sizes
    """
    soms = []

    for map_size in map_sizes:
        print(f"Training SOM with map size {map_size}...")
        som = SelfOrganizingMap(
            map_size=map_size,
            input_dim=input_dim,
            learning_rate=learning_rate,
            sigma=sigma,
            neighborhood_function=neighborhood_function,
            n_iterations=n_iterations
        )
        som.train(data, verbose=False)
        soms.append(som)

    return soms

# Train SOMs with different map sizes
map_sizes = [(5, 5), (10, 10), (15, 15), (20, 20)]
soms_different_sizes = train_soms_with_different_sizes(
    X_iris_scaled, 4, map_sizes, 0.5, 2.0, 'gaussian', 5000
)

# Visualize sample mappings for different map sizes
fig, axes = plt.subplots(1, len(map_sizes), figsize=(20, 5))

for i, (som, map_size) in enumerate(zip(soms_different_sizes, map_sizes)):
    som.visualize_sample_mapping(X_iris_scaled, y_iris, ax=axes[i],
                               title=f"Map Size: {map_size[0]}x{map_size[1]}")

plt.tight_layout()
plt.show()

# Calculate error metrics for different map sizes
topographic_errors_size = []
quantization_errors_size = []

for som in soms_different_sizes:
    te = calculate_topographic_error(som, X_iris_scaled)
    qe = calculate_quantization_error(som, X_iris_scaled)
    topographic_errors_size.append(te)
    quantization_errors_size.append(qe)

# Plot error metrics vs map size
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Convert map sizes to total number of neurons for plotting
map_sizes_neurons = [size[0] * size[1] for size in map_sizes]

# Plot topographic error vs map size
axes[0].plot(map_sizes_neurons, topographic_errors_size, 'o-', color='blue')
axes[0].set_xlabel('Number of Neurons')
axes[0].set_ylabel('Topographic Error')
axes[0].set_title('Topographic Error vs Map Size')
axes[0].grid(True)

# Plot quantization error vs map size
axes[1].plot(map_sizes_neurons, quantization_errors_size, 'o-', color='red')
axes[1].set_xlabel('Number of Neurons')
axes[1].set_ylabel('Quantization Error')
axes[1].set_title('Quantization Error vs Map Size')
axes[1].grid(True)

plt.tight_layout()
plt.show()

# Part 11: Conclusion and key findings

print("\nKey Findings from the SOM Tutorial:")
print("1. Neighborhood Function Impact:")
print("   - Gaussian: Smooth, gradual weight adaptation with good topological preservation")
print("   - Mexican Hat: Creates more distinct clusters but may have higher topographic error")
print("   - Bubble: Simpler adaptation with more defined boundaries but potentially less smooth topology")

print("\n2. Parameter Influence:")
print("   - Learning rate affects the speed and stability of convergence")
print("   - Sigma (neighborhood radius) controls the spread of adaptation influence")
print("   - Map size impacts resolution and quantization error")

print("\n3. SOM vs t-SNE:")
print("   - SOM provides a structured grid layout preserving topology")
print("   - t-SNE focuses on preserving local similarities but loses global structure")
print("   - SOM offers better interpretability through its regular grid")

print("\n4. Convergence Behavior:")
print("   - Quantization error generally decreases with more iterations")
print("   - Different neighborhood functions show unique convergence patterns")
print("   - Training progress visualizations show how the SOM adapts over time")

print("\n5. Practical Applications:")
print("   - Visualization of high-dimensional data (e.g., Iris dataset, faces)")
print("   - Cluster identification and boundary analysis through U-matrix")
print("   - Topological preservation of input space relationships")

# References
print("\nReferences:")
print("1. Kohonen, T. (1990). The self-organizing map. Proceedings of the IEEE, 78(9), 1464-1480.")
print("2. Vesanto, J., & Alhoniemi, E. (2000). Clustering of the self-organizing map. IEEE Transactions on Neural Networks, 11(3), 586-600.")
print("3. Ultsch, A., & Siemon, H. P. (1990). Kohonen's self organizing feature maps for exploratory data analysis. In Proceedings of the International Neural Network Conference (pp. 305-308).")
print("4. Plzlbauer, G. (2004). Survey and comparison of quality measures for self-organizing maps. In Proceedings of the Fifth Workshop on Data Analysis (WDA'04) (pp. 67-82).")
print("5. Van Der Maaten, L., & Hinton, G. (2008). Visualizing data using t-SNE. Journal of machine learning research, 9(11).")
```
